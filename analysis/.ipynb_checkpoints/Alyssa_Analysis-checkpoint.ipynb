```python
import sys
import os
import glob
import importlib
#
import numpy as np
import pylab as plt
import matplotlib
import math as m
from matplotlib import rc
import sys
from scipy.optimize import curve_fit

#
def findFFT(listofIDs,IX,subject):
    #sys.path.insert(0,'C:\\Users\\BRL\\Momona\'s Google Drive\\Yamagami Lab_\\Slider Project\\hcps\\develop\\')
    trials = dict()
    for id in listofIDs:
        
      fis = sorted(glob.glob(os.path.join('C:\\Users\\BRL\\Momona\'s Google Drive\\Yamagami Lab_\\Slider Project\\hcps\\experiment\\data\\',subject,id+'.npz')))#'*_'+protocol+'_'+id+'.npz')))
      if len(fis) > 1:
        dbg('WARNING -- repeated trials for id ='+id)
      assert len(fis) > 0, 'ERROR -- no data for id ='+id
      fi = fis[-1]
      #print('LOAD '+fi)
      trial = dict(np.load(fi,encoding="latin1"))
      trials[id] = trial

    timedomainvalues = {}
    timedomainvalues[listofIDs[0]] = {}
    times_ = [trials[listofIDs[0]]['time_']]
    refs_  = [trials[listofIDs[0]]['ref_']]
    outs_  = [trials[listofIDs[0]]['out_']]
    inps_  = [trials[listofIDs[0]]['inp_']]#/(scales[0]*3.)] # what's the scale for?
    dists_ = [trials[listofIDs[0]]['dis_']]


    timedomainvalues[listofIDs[0]]['times'] = np.hstack(times_)[-2400:] # take out first 5 sec
    timedomainvalues[listofIDs[0]]['refs'] = np.hstack(refs_)[-2400:]
    timedomainvalues[listofIDs[0]]['outs'] = np.hstack(outs_)[-2400:]
    timedomainvalues[listofIDs[0]]['inps'] = np.hstack(inps_)[-2400:]
    timedomainvalues[listofIDs[0]]['dists'] = np.hstack(dists_)[-2400:]

    for id in listofIDs[1:]:
      times_ = [trials[id]['time_']]
      refs_  = [trials[id]['ref_']]
      outs_  = [trials[id]['out_']]
      inps_  = [trials[id]['inp_']]
      dists_ = [trials[id]['dis_']]
      timedomainvalues[id] = {}
      timedomainvalues[id]['times'] = (np.hstack(times_)[-2400:]) # take out first 5 sec
      timedomainvalues[id]['refs']=(np.hstack(refs_)[-2400:])
      timedomainvalues[id]['outs']=(np.hstack(outs_)[-2400:])
      timedomainvalues[id]['inps']=(np.hstack(inps_)[-2400:])
      timedomainvalues[id]['dists']=(np.hstack(dists_)[-2400:])


    FREQDOMAINVALUES = {}
    for id in listofIDs:
        FREQDOMAINVALUES[id] = {}
        FREQDOMAINVALUES[id]['OUTS'] = np.fft.fft(timedomainvalues[id]['outs'])[:int(timedomainvalues[id]['outs'].shape[0]/2)]/TEMPFREQ
        FREQDOMAINVALUES[id]['INPS'] = np.fft.fft(timedomainvalues[id]['inps'])[:int(timedomainvalues[id]['inps'].shape[0]/2)]/TEMPFREQ
        FREQDOMAINVALUES[id]['REFS'] = np.fft.fft(timedomainvalues[id]['refs'])[:int(timedomainvalues[id]['refs'].shape[0]/2)]/TEMPFREQ
        FREQDOMAINVALUES[id]['DISTS'] = np.fft.fft(timedomainvalues[id]['dists'])[:int(timedomainvalues[id]['dists'].shape[0]/2)]/TEMPFREQ
        FREQDOMAINVALUES[id]['Hru'] = np.divide(FREQDOMAINVALUES[id]['INPS'][IX],FREQDOMAINVALUES[id]['REFS'][IX])
        FREQDOMAINVALUES[id]['Hdu'] = np.divide(FREQDOMAINVALUES[id]['INPS'][IX],FREQDOMAINVALUES[id]['DISTS'][IX])
    return timedomainvalues, FREQDOMAINVALUES

def exponential_func(x,a,b,c):
    return a*np.exp(-b*x)+c

BOTH = []
def plotThingsEO(M,Bothid,Evenid,Oddid,IX,IXE,IXO,subject):
    err = np.zeros((int(len(Bothid)-1),),dtype=complex)
    timeerr = np.zeros((int(len(Bothid)-1),))
    
    # find ffts
    global BOTH
    both,BOTH = findFFT(Bothid,IX,subject)
    
    # TODO TAKE OUT
    plt.figure()
    plt.semilogx(abs(BOTH[Bothid[0]]['REFS']),label='Reference')
    plt.semilogx(abs(BOTH[Bothid[0]]['DISTS']),label='Disturbance')
    plt.legend()
    
    plt.figure()
    plt.semilogx(abs(BOTH[Bothid[1]]['REFS']),label='Reference')
    plt.semilogx(abs(BOTH[Bothid[1]]['DISTS']),label='Disturbance')
    plt.legend()
    
    
    BA = np.zeros((int(len(Bothid)-1),len(IX)),dtype=np.complex_)
    FA = np.zeros((int(len(Bothid)-1),len(IX)),dtype=np.complex_)
    count = 0
    for num,id in enumerate(Bothid[:-1]):
        timeerr[num] = np.mean((np.asarray(both[id]['outs'])-np.asarray(both[id]['refs']))**2)

        firstID = id
        secID = Bothid[num+1]
        Hru = np.zeros((len(IX),),dtype=complex)
        Hdu = np.zeros((len(IX),),dtype=complex)
        if firstID in Evenid:
            Hru[0::2] = BOTH[firstID].get('Hru')[0::2]
            Hru[1::2] = BOTH[secID].get('Hru')[1::2]
            Hdu[1::2] = BOTH[firstID].get('Hdu')[1::2]
            Hdu[0::2] = BOTH[secID].get('Hdu')[0::2]
        elif firstID in Oddid:
            Hru[0::2] = BOTH[secID].get('Hru')[0::2]
            Hru[1::2] = BOTH[firstID].get('Hru')[1::2]
            Hdu[1::2] = BOTH[secID].get('Hdu')[1::2]
            Hdu[0::2] = BOTH[firstID].get('Hdu')[0::2]
        else:
            print('error: no ID available')
        if num == 0:
            plt.figure()
            plt.semilogx(abs(BOTH[firstID]['REFS']),label='refs')
            plt.semilogx(abs(BOTH[firstID]['DISTS']),label='dists')
            plt.semilogx(abs(BOTH[firstID]['INPS']),label='inps')
            plt.legend()
        BA[count,:] = np.divide(-Hdu,(np.multiply(M,(1.0+0j)+Hdu)))
        FA[count,:] = np.multiply(Hru,(1.0+0j)+np.multiply(BA[count,:],M))-BA[count,:]
        count+=1
    plt.figure(figsize=(20,10))
    plt.subplot(221)
    num = len(Bothid)/2
    linecolor = 0.1 # darker = smaller number; lighter = larger number closer to 1
    
    for idNum in range(0,len(BA),1):
        temperr = 0+0j
        err[idNum] = np.mean(((abs(FA[idNum]-np.divide(1,M)))**2))

        plt.loglog(freq[IX],np.absolute(FA[idNum]),color='%f'%linecolor)#,label='Feedforward')
        linecolor += 0.05
    plt.loglog(freq[IX],np.absolute(np.divide(1,M)),'r--',label='Model Inverse Prediction')
    plt.legend()
    plt.ylabel('Gain')

    linecolor = 0.2
    plt.subplot(223)
    for idNum in range(0,len(BA),1):
        #plt.semilogx(freq[IX],np.angle(BA[idNum])*180/m.pi,'b')#,label='Feedback')
        plt.semilogx(freq[IX],np.angle(FA[idNum])*180/m.pi,color='%f'%linecolor)#,label='Feedforward')
        # CENTER THE GRAPH AROUND 1/M
        linecolor += 0.05
    plt.semilogx(freq[IX],np.angle(np.divide(1,M))*180/m.pi,'r--',label='Model Inverse Prediction')
    plt.legend()
    plt.ylabel('Phase - deg')
    plt.xlabel('Freq - Hz')

#     plt.figure()
#     plt.loglog(IX,np.absolute(BAavg),'o')
#     plt.title('BA average magnitude')
#     plt.legend()

    # CURVE FIT TO ERROR
    # define function (logarithmic)
    xx = np.linspace(0,len(err),num=len(err))
    yy = np.absolute(err)
    
    def sigmoid(x, x0, k,a,b):
        y = a / (b + np.exp(-k*(x-x0)))
        return y
    popt,pcov = curve_fit(sigmoid,xx,yy)

    #popt,pcov = curve_fit(func,np.linspace(0,len(err),num=len(err)),np.absolute(err),p0=(-.1,-.1,-.1))
    print(popt)
    
    plt.figure()
    plt.plot(np.linspace(0,len(err)-1,num=len(err)),np.absolute(err),'o',label='Raw Data')
    plt.plot(np.linspace(0,len(err)-1,num=len(err)),sigmoid(np.absolute(err),*popt),'r--',label='Fitted curve')
    plt.legend()
    plt.title('FF-Minv err magnitude')

    
    plt.figure()
    plt.plot(np.linspace(0,len(err)-1,num=len(err)),np.angle(err),'o')
    plt.title('FF-Minv err phase')
    
    plt.figure()
    plt.plot(np.linspace(0,len(timeerr)-1,num=len(timeerr)),timeerr,'o')
    plt.title('MSE')
    
    print(np.mean(err))
    print(err)
    
#     plt.figure()
#     plt.plot(np.linspace(0,39,num=39),timeerr[:-1],'o')
#     print(timeerr)
#     plt.title('ref-y vs num trials')
help = """
usage:
  analysis subject protocol [formats]

add filename formats to save files in those formats:
  visualization subject protocol png,mp4
"""
# importing protocol (su17v3)
sys.path.append('C:\\Users\\BRL\\Momona\'s Google Drive\\Yamagami Lab_\\Slider Project\\hcps\\experiment\\protocols')
import su18CP as proto
#sys.path.append('C:\\Users\\BRL\\Momona\'s Google Drive\\Yamagami Lab_\\Slider Project\\hcps\\develop')

# find files for specific subject
def findFilename(subject):
    fis = glob.glob('C:\\Users\\BRL\\Momona\'s Google Drive\\Yamagami Lab_\\Slider Project\\hcps\\experiment\\data/'+subject+'\\*.npz')
    #fis = glob.glob('C:\\Users\\BRL\\Momona\'s Google Drive\\Yamagami Lab_\\Slider Project\\hcps\\develop\\data/'+subject+'\\*.npz')
    #print(fis)
    fis = [os.path.basename(fi) for fi in fis]

    ids = sorted(list(set([fi.strip('.npz') for fi in fis])))

    #ids = [id for id in ids if ('.csv') not in id[-4:]]
    ids = [id for id in ids if ('st2') not in id]
    ids = [id for id in ids if ('st1') not in id]
    ids = [id for id in ids if ('st0') not in id]
    #ids = [id for id in ids if ('st2', 'st1','rej','oob','rst','max','st1','st2') not in id]
    #ids = [id for id in ids if ('so') in id]
    Aid = [id for id in ids if ('r-sos+A_d-sos+A') in id]
    Evenid = [id for id in ids if ('r-sos+E') in id or ('r-sos-E') in id]
    Oddid = [id for id in ids if ('r-sos-O') in id or ('r-sos+O') in id]
    Bothid = [id for id in ids if ('sos+A') not in id]
    return ids,Evenid,Oddid
# Define
primes = np.asarray([2, 3, 5, 7, 11, 13,17,19])# max =37
IX = primes*2
primesE = np.asarray([2,5,11,17])#primes[::2]
primesO = np.asarray([3,7,13,19])#primes[1::2]
IXE = primesE*2
IXO = primesO*2
freq = np.linspace(0,599.5,1200)#refs.shape[0]/2)
tempfreq = np.ones((2400,))
TEMPFREQ = np.fft.fft(tempfreq)[0]
s = 1.0j*2*m.pi*freq[IX]/20
M = 1.0/(np.multiply(s,s)+s)
M = np.asarray(M)
```
